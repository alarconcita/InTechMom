```python
import pandas as pd
import mysql.connector
from mysql.connector import Error
```

```python
df = pd.read_csv('/content/drive/MyDrive/michelin_my_maps.csv')
```

```python
print(df.head())
```

```python
print(df.columns)
```

```python
# Agregar columnas 'Country' y 'Locality' dividiendo la columna 'Location'
df[['Locality', 'Country']] = df['Location'].str.rsplit(', ', n=1, expand=True)
print(df.columns)
df.head()
```

```python
df.Price.unique()
```

```python
# Columna 'Price'
currency_symbols = ['¥', '£', '€', '$', '₩', '₺', '฿', '₫']
df['Price'] = df['Price'].replace({symbol: '' for symbol in currency_symbols})

df['Price'] = df['Price'].apply(lambda x: 'Muy Barato' if pd.isna(x) or len(str(x)) == 0
                                  else 'Accesible' if len(str(x)) == 1
                                  else 'Normal' if len(str(x)) == 2
                                  else 'Costoso' if len(str(x)) == 3
                                  else 'Muy costoso' if len(str(x)) >= 4
                                  else x)

print(df[['Name', 'Price']].head())
```

```python
# Dividir 'Cuisine'
max_cuisine_labels = df['Cuisine'].str.count(',').max() + 1 if not pd.isna(df['Cuisine']).all() else 1
cuisine_columns = [f'Cuisine {i}' for i in range(1, max_cuisine_labels + 1)]
df[cuisine_columns] = df['Cuisine'].str.split(', ', expand=True)

print(df.columns)
df.head()
```

```python
# Filtrar por columnas 'Award'
df = df[df['Award'].isin(['1 Start', '2 Starts', '3 Starts'])]
```

```python
# Establecer un máximo deseado de etiquetas
max_facilities_labels = 5

# Dividir 'FacilitiesAndServices' y expandir en columnas
split_columns = df['FacilitiesAndServices'].str.split(', ', n=max_facilities_labels - 1, expand=True)

# Concatenar las nuevas columnas al DataFrame original
df = pd.concat([df, split_columns], axis=1)

# Eliminar la columna original 'FacilitiesAndServices' si es necesario
df = df.drop('FacilitiesAndServices', axis=1)

# Renombrar las nuevas columnas
df.columns = df.columns[:len(df.columns)-max_facilities_labels].tolist() + [f'Facilities_{i}' for i in range(1, max_facilities_labels + 1)]

print(df.columns)
```

```python
# Imprimir el DataFrame resultante
print(df.head())
```
